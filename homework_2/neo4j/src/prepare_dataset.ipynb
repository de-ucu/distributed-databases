```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/raw.csv.xz")
df
```

```python
items = df[["StockCode", "Description", "UnitPrice"]]
items = items.drop_duplicates(subset=["StockCode"], ignore_index=True)
items.to_csv("../data/items.csv.xz", index=False)
items
```

```python
users = pd.Series(df.CustomerID.unique())
users.to_csv("../data/users.csv.xz", index=False)
users
```

```python
transactions = pd.Series(df.InvoiceNo.unique())
transactions.to_csv("../data/transactions.csv.xz", index=False)
transactions
```

```python
orders = df[["InvoiceNo", "StockCode", "Quantity", "CustomerID"]]
orders.to_csv("../data/orders.csv.xz", index=False)
orders
```

```python
views_count = 80000

np.random.seed(1)
probabilities = np.random.exponential(size=users.shape[0])
probabilities
probabilities /= sum(probabilities)
viewers = np.random.choice(users, views_count, p=probabilities)

np.random.seed(2)
probabilities = np.random.lognormal(size=items.shape[0])
probabilities
probabilities /= sum(probabilities)
viewed_items = np.random.choice(items.StockCode, views_count, p=probabilities)

views = pd.DataFrame({
    "CustomerID": viewers,
    "StockCode": viewed_items,
})
views = views.drop_duplicates(ignore_index=True)
views.to_csv("../data/views.csv.xz", index=False)
views
```
